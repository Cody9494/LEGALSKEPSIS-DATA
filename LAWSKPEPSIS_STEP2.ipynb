```python
import re
import pandas as pd

JSON_COL = "content_json"   # <-- Œ¨ŒªŒªŒ±ŒæŒµ ŒµŒ¥œé Œ±ŒΩ Œ∑ œÉœÑŒÆŒªŒ∑ œÉŒøœÖ Œ≠œáŒµŒπ Œ¨ŒªŒªŒø œåŒΩŒøŒºŒ±

def extract_text(obj):
    """Return readable text from str/dict/list. Keeps only #text for dict by default."""
    if obj is None:
        return ""
    if isinstance(obj, str):
        return obj
    if isinstance(obj, dict):
        # very common in Formex: {"#text": "...", "DATE": {...}, "HT": {...}}
        return obj.get("#text", "")
    if isinstance(obj, list):
        return " ".join(extract_text(x) for x in obj)
    return ""

def flatten_text(obj):
    """
    Deeper flatten: concatenates all nested #text strings in dict/list,
    and also supports keys like DATE/HT/NOTE/QUOT.* without losing main text.
    """
    if obj is None:
        return ""
    if isinstance(obj, str):
        return obj
    if isinstance(obj, list):
        parts = [flatten_text(x) for x in obj]
        return " ".join([p for p in parts if p]).strip()
    if isinstance(obj, dict):
        parts = []
        # prioritize #text
        if "#text" in obj and isinstance(obj["#text"], str):
            parts.append(obj["#text"])
        # then walk other values
        for k, v in obj.items():
            if k == "#text":
                continue
            parts.append(flatten_text(v))
        return " ".join([p for p in parts if p]).strip()
    return ""
```

```python
import json

def ensure_dict(x):
    """
    Convert x to dict if possible:
    - dict -> as-is
    - str  -> json.loads (or return {} if invalid)
    - else -> {}
    """
    if isinstance(x, dict):
        return x
    if isinstance(x, str):
        s = x.strip()
        if not s:
            return {}
        try:
            return json.loads(s)
        except json.JSONDecodeError:
            # ŒëŒΩ Œ≠œáŒµŒπœÇ single quotes ŒÆ Œ∫Œ¨œÑŒπ "œÉœÄŒ±œÉŒºŒ≠ŒΩŒø", Œ∫œÅŒ¨œÑŒ± Œ∫ŒµŒΩœå Œ≥ŒπŒ± ŒΩŒ± œÑŒø ŒµŒΩœÑŒøœÄŒØœÉŒøœÖŒºŒµ
            return {}
    return {}
```

```python
def parse_header_fields(content_dict: dict) -> dict:
    act = (content_dict or {}).get("ACT") or {}

    # act_id
    bib = act.get("BIB.INSTANCE") or {}
    no_doc = bib.get("NO.DOC") or {}
    if isinstance(no_doc, list):
        no_doc = no_doc[0] if no_doc else {}
    year = (no_doc or {}).get("YEAR", "")
    number = (no_doc or {}).get("NO.CURRENT", "")
    act_id = f"{year}/{number}" if year and number else ""

    # title + act_type + reg_number
    title_parts = (((act.get("TITLE") or {}).get("TI") or {}).get("P")) or []
    if isinstance(title_parts, dict):
        title_parts = [title_parts]
    title_str = " ".join(flatten_text(p) for p in title_parts).strip()

    reg_match = re.search(r"\d{4}/\d+", title_str)
    reg_number = act_id if reg_match else ""

    act_type = ""
    if title_parts:
        first = title_parts[0]
        if isinstance(first, dict) and "HT" in first:
            act_type = flatten_text(first["HT"]).strip()

    # date_signed
    date_elem = bib.get("DATE") or {}
    if isinstance(date_elem, list):
        date_elem = date_elem[0] if date_elem else {}
    date_signed = (date_elem or {}).get("@ISO", "")

    # legal_basis (GR.VISA)
    legal_basis = []
    preamble = act.get("PREAMBLE") or {}
    gr_visa = preamble.get("GR.VISA") or {}
    gr_visa_list = gr_visa if isinstance(gr_visa, list) else [gr_visa]
    for gv in gr_visa_list:
        if not isinstance(gv, dict):
            continue
        visas = gv.get("VISA", [])
        if isinstance(visas, (dict, str)):
            visas = [visas]
        elif visas is None:
            visas = []
        for v in visas:
            t = flatten_text(v).strip()
            if t:
                legal_basis.append(t)

    # recitals (flat + nested DIV structures)
    recitals = []

    def collect_recitals(node):
        """Collect any NP blocks under GR.CONSID and nested DIV.CONSID..."""
        if node is None:
            return
        if isinstance(node, list):
            for x in node:
                collect_recitals(x)
            return
        if isinstance(node, dict):
            # direct CONSIDs
            if "CONSID" in node:
                collect_recitals(node["CONSID"])
            # NP blocks
            if "NP" in node:
                np = node["NP"]
                if isinstance(np, list):
                    for npi in np:
                        collect_recitals({"NP": npi})
                elif isinstance(np, dict):
                    txt = np.get("TXT", "")
                    t = flatten_text(txt).strip()
                    if t:
                        recitals.append(t)
            # recurse everything else
            for _, v in node.items():
                collect_recitals(v)

    gr_consid = preamble.get("GR.CONSID") or {}
    collect_recitals(gr_consid)

    # signatory
    signatory = []
    signature = ((act.get("FINAL") or {}).get("SIGNATURE")) or {}
    if isinstance(signature, dict):
        sig = signature.get("SIGNATORY", [])
        if isinstance(sig, dict):
            sig = [sig]
        elif not isinstance(sig, list):
            sig = []
        for s in sig:
            if not isinstance(s, dict):
                continue
            ps = s.get("P", [])
            if isinstance(ps, (dict, str)):
                ps = [ps]
            elif ps is None:
                ps = []
            for p in ps:
                t = flatten_text(p).strip()
                if t:
                    signatory.append(t)

    return {
        "act_id": act_id,
        "act_type": act_type,
        "reg_number": reg_number,
        "title": title_str,
        "date_signed": date_signed,
        "legal_basis": legal_basis,
        "recitals": recitals,
        "signatory": signatory,
    }
```

```python
def normalize_alinea_text(alinea_obj):
    """
    - flattens text
    - fixes patterns like "From ," using DATE.#text when present
    - fixes "in the ." when HT exists (Official Journal...)
    """
    if alinea_obj is None:
        return ""

    # If dict with DATE + #text
    if isinstance(alinea_obj, dict):
        main = alinea_obj.get("#text", "")
        date_txt = ""
        if "DATE" in alinea_obj:
            date_txt = flatten_text(alinea_obj["DATE"]).strip()

        ht_txt = ""
        if "HT" in alinea_obj:
            ht_txt = flatten_text(alinea_obj["HT"]).strip()

        # fallback flatten
        if not isinstance(main, str) or not main.strip():
            main = flatten_text(alinea_obj)

        if isinstance(main, str):
            if date_txt and "From ," in main:
                main = main.replace("From ,", f"From {date_txt},")
            if date_txt and "on ." in main:
                main = main.replace("on .", f"on {date_txt}.")
            if ht_txt and "in the ." in main:
                main = main.replace("in the .", f"in the {ht_txt}.")
        return (main or "").strip()

    # list of alinea blocks
    if isinstance(alinea_obj, list):
        parts = [normalize_alinea_text(x) for x in alinea_obj]
        return " ".join([p for p in parts if p]).strip()

    # string
    if isinstance(alinea_obj, str):
        return alinea_obj.strip()

    return flatten_text(alinea_obj).strip()


def parse_articles_from_act(content_dict: dict):
    act = (content_dict or {}).get("ACT") or {}
    terms = act.get("ENACTING.TERMS") or {}
    if not isinstance(terms, dict):
        terms = {}

    articles_out = []

    def parse_article_node(a):
        if not isinstance(a, dict):
            return
        num = flatten_text(a.get("TI.ART", "")).replace("Article", "").strip()
        title = flatten_text(a.get("STI.ART", "")).strip() if "STI.ART" in a else ""

        body_parts = []

        # Prefer PARAG[] if present
        parag = a.get("PARAG", None)
        if parag is not None:
            parags = parag if isinstance(parag, list) else [parag]
            for pg in parags:
                if not isinstance(pg, dict):
                    continue
                no_parag = flatten_text(pg.get("NO.PARAG", "")).strip()
                alinea = normalize_alinea_text(pg.get("ALINEA", ""))
                if alinea:
                    if no_parag:
                        body_parts.append(f"{no_parag} {alinea}".strip())
                    else:
                        body_parts.append(alinea)

        # Also include ARTICLE-level ALINEA
        alinea_direct = a.get("ALINEA", None)
        if alinea_direct is not None:
            alinea_txt = normalize_alinea_text(alinea_direct)
            if alinea_txt:
                body_parts.append(alinea_txt)

        body = " ".join(body_parts).strip()

        articles_out.append({
            "article_number": num,
            "article_title": title,
            "article_body": body
        })

    # Case 1: terms.ARTICLE
    arts = terms.get("ARTICLE", [])
    if isinstance(arts, dict):
        arts = [arts]
    if isinstance(arts, list):
        for a in arts:
            parse_article_node(a)

    # Case 2: terms.DIVISION[].ARTICLE[]
    divs = terms.get("DIVISION", [])
    if isinstance(divs, dict):
        divs = [divs]
    if isinstance(divs, list):
        for d in divs:
            if not isinstance(d, dict):
                continue
            div_arts = d.get("ARTICLE", [])
            if isinstance(div_arts, dict):
                div_arts = [div_arts]
            if isinstance(div_arts, list):
                for a in div_arts:
                    parse_article_node(a)

    return articles_out
```

```python
def build_normalized_text(row):
    # Œ∫œÅŒ±œÑŒ¨ŒºŒµ structured fields, Œ±ŒªŒªŒ¨ ŒîŒïŒù œÄŒµœÑŒ¨ŒºŒµ raw json
    title = row.get("title","") or ""
    lb = row.get("legal_basis", []) or []
    rec = row.get("recitals", []) or []
    arts = row.get("articles", []) or []
    sig = row.get("signatory", []) or []

    arts_txt = []
    for a in arts:
        if not isinstance(a, dict):
            continue
        n = a.get("article_number","")
        t = a.get("article_title","")
        b = a.get("article_body","")
        chunk = f"Article {n} ‚Äî {t}. {b}".strip()
        if chunk:
            arts_txt.append(chunk)

    parts = [
        title.strip(),
        ("LEGAL BASIS: " + " ".join(lb)).strip() if lb else "",
        ("RECITALS: " + " ".join(rec)).strip() if rec else "",
        ("ARTICLES: " + " ".join(arts_txt)).strip() if arts_txt else "",
        ("SIGNATORY: " + " ".join(sig)).strip() if sig else "",
    ]
    return "\n".join([p for p in parts if p]).strip()


def build_df_master(df_act: pd.DataFrame) -> pd.DataFrame:
    rows = []
    for _, r in df_act.iterrows():
        d = r[JSON_COL]
        parsed = parse_header_fields(d)
        parsed["articles"] = parse_articles_from_act(d)

        out = {
            "filename": r.get("filename", None),
            "filepath": r.get("filepath", None),
            **parsed,
        }
        out["normalized_text"] = build_normalized_text(out)

        # keep raw json reference if you want (recommended)
        out["content_json"] = d  # optional but safest

        rows.append(out)

    return pd.DataFrame(rows)
```

```python
def count_bad_placeholders(text: str) -> dict:
    if not isinstance(text, str):
        text = ""
    return {
        "from_comma": text.count("From ,"),
        "in_the_dot": text.count("in the ."),
        "on_dot": text.count("on ."),
        "from_dot": text.count("from ."),
    }
```

```python
def looks_truncated(s: str) -> bool:
    if not isinstance(s, str) or not s.strip():
        return False
    s = s.strip()
    # suspicious endings: cut mid-word, or no punctuation for long sentence
    if len(s) > 80 and re.search(r"[A-Za-z]{3,}$", s) and not s.endswith((".", ")", "]", "‚Äô", "‚Äù", '"', ";", ":")):
        return True
    # classic cut marker: ends with very short tail + no punctuation
    if len(s) > 120 and not re.search(r"[.!?]$", s):
        # not always wrong, but flag
        return True
    return False
```

```python
def validate_df_master(df_master: pd.DataFrame) -> pd.DataFrame:
    bads = df_master["normalized_text"].apply(count_bad_placeholders).apply(pd.Series)
    dfv = pd.concat([df_master[["filename","act_id"]], bads], axis=1)

    # truncation per article
    trunc_rows = []
    for _, r in df_master.iterrows():
        for a in (r.get("articles") or []):
            body = (a or {}).get("article_body","")
            if looks_truncated(body):
                trunc_rows.append({
                    "filename": r.get("filename"),
                    "act_id": r.get("act_id"),
                    "article_number": (a or {}).get("article_number"),
                    "tail": body[-80:]
                })
    trunc_df = pd.DataFrame(trunc_rows)
    return dfv, trunc_df
```

```python
def build_df_master(df_act: pd.DataFrame) -> pd.DataFrame:
    rows = []
    for _, r in df_act.iterrows():
        raw = r[JSON_COL]
        d = ensure_dict(raw)   # <-- ŒµŒ¥œé œÑŒø fix

        parsed = parse_header_fields(d)
        parsed["articles"] = parse_articles_from_act(d)

        out = {
            "filename": r.get("filename", None),
            "filepath": r.get("filepath", None),
            **parsed,
        }
        out["normalized_text"] = build_normalized_text(out)

        # keep raw + parsed dict for safety
        out["content_json_raw"] = raw
        out["content_json"] = d

        rows.append(out)

    return pd.DataFrame(rows)
```

```python
import re

def normalize_ws(s: str) -> str:
    s = re.sub(r"\s+", " ", (s or "")).strip()
    return s

def dedup_preserve_order(items):
    seen = set()
    out = []
    for x in items or []:
        nx = normalize_ws(x)
        if not nx:
            continue
        key = nx.casefold()
        if key in seen:
            continue
        seen.add(key)
        out.append(nx)
    return out
```

```python
NOISE_PATTERNS = [
    r"\bITALIC\b", r"\bUC\b", r"\bARAB\b", r"\bFOOTNOTE\b",
    r"\bQS\d+\b", r"\bQE\d+\b", r"\bE\d{4}\b",   # E0001 etc
]

def strip_noise_tokens(s: str) -> str:
    if not s:
        return s
    out = s
    for pat in NOISE_PATTERNS:
        out = re.sub(pat, "", out)
    out = re.sub(r"\s+", " ", out).strip()
    out = re.sub(r"\s+([.,;:])", r"\1", out)  # fix spaces before punctuation
    return out
```

```python
df_master["title"] = df_master["title"].apply(strip_noise_tokens)
df_master["act_type"] = df_master["act_type"].apply(strip_noise_tokens)

df_master["legal_basis"] = df_master["legal_basis"].apply(lambda xs: [strip_noise_tokens(x) for x in (xs or [])])
df_master["recitals"] = df_master["recitals"].apply(lambda xs: [strip_noise_tokens(x) for x in (xs or [])])

def clean_articles(arts):
    out = []
    for a in arts or []:
        out.append({
            **a,
            "article_title": strip_noise_tokens(a.get("article_title","")),
            "article_body": strip_noise_tokens(a.get("article_body","")),
        })
    return out

df_master["articles"] = df_master["articles"].apply(clean_articles)
```

```python
import json
import pandas as pd

def as_str(x):
    if x is None or (isinstance(x, float) and pd.isna(x)):
        return ""
    return str(x)

def as_list_str(x):
    if x is None or (isinstance(x, float) and pd.isna(x)):
        return []
    if isinstance(x, list):
        return [as_str(i).strip() for i in x if as_str(i).strip()]
    # Œ±ŒΩ Œ≠œÅŒ∏ŒµŒπ string (œÄ.œá. Œ±œÄœå csv reload), Œ∫œÅŒ¨œÑŒ± œÑŒø œâœÇ 1 œÉœÑŒøŒπœáŒµŒØŒø
    s = as_str(x).strip()
    return [s] if s else []

def as_articles(x):
    # Œ∏Œ≠ŒªŒøœÖŒºŒµ list[dict]
    if x is None or (isinstance(x, float) and pd.isna(x)):
        return []
    if isinstance(x, list):
        out = []
        for a in x:
            if not isinstance(a, dict):
                continue
            out.append({
                "article_number": as_str(a.get("article_number")).strip(),
                "article_title": as_str(a.get("article_title")).strip(),
                "article_body":  as_str(a.get("article_body")).strip(),
            })
        return out
    # Œ±ŒΩ Œ±œÄœå csv reload Œ≠Œ≥ŒπŒΩŒµ string, œÄœÅŒøœÉœÄŒ¨Œ∏Œ∑œÉŒµ json
    if isinstance(x, str):
        try:
            obj = json.loads(x)
            return as_articles(obj)
        except Exception:
            return []
    return []
```

```python
import re

def _safe_first(x):
    if isinstance(x, list):
        return x[0] if x else {}
    return x

def extract_act_id(content_dict, filename=None):
    """
    Returns act_id like '2025/1003' or ''.
    content_dict must be a dict with key 'ACT' (or already the act dict).
    """
    if not isinstance(content_dict, dict):
        return ""

    act = content_dict.get("ACT") if "ACT" in content_dict else content_dict
    if not isinstance(act, dict):
        return ""

    bib = act.get("BIB.INSTANCE") or {}

    # (1) Preferred: BIB.INSTANCE.NO.DOC
    no_doc = _safe_first(bib.get("NO.DOC") or {})
    if isinstance(no_doc, dict):
        year = (no_doc.get("YEAR") or "").strip()
        number = (no_doc.get("NO.CURRENT") or "").strip()
        if year and number:
            return f"{year}/{number}"

    # (2) Fallback: BIB.INSTANCE.DOCUMENT.REF.NO.DOC
    doc_ref = _safe_first(bib.get("DOCUMENT.REF") or {})
    no_doc2 = doc_ref.get("NO.DOC") if isinstance(doc_ref, dict) else None

    # NO.DOC ŒºœÄŒøœÅŒµŒØ ŒΩŒ± ŒµŒØŒΩŒ±Œπ dict ŒÆ string
    if isinstance(no_doc2, dict):
        year = (no_doc2.get("YEAR") or "").strip()
        number = (no_doc2.get("NO.CURRENT") or "").strip()
        if year and number:
            return f"{year}/{number}"
    elif isinstance(no_doc2, str):
        # œÄ.œá. "... YEAR 2012 ... NO.CURRENT 187 ..."
        m = re.search(r"\b(19|20)\d{2}\b.*?\b(\d{1,5})\b", no_doc2)
        if m:
            return f"{m.group(0)}/{m.group(1)}"  # Œ±ŒΩ Œ≤Œ≥ŒµŒπ œÄŒµœÅŒØŒµœÅŒ≥Œø, Œ±œÄŒªŒ¨ œÑŒø Œ±Œ≥ŒΩŒøŒµŒØœÇ

    # (3) Fallback: regex from title text
    title_parts = act.get("TITLE", {}).get("TI", {}).get("P", [])
    if isinstance(title_parts, dict):
        title_parts = [title_parts]
    title_str = " ".join(extract_text(p) for p in title_parts).strip()
    m = re.search(r"\b(19|20)\d{2}/\d{1,5}\b", title_str)
    if m:
        return m.group(0)

    # (4) Fallback: regex from filename (L_202501003EN... => 2025/1003)
    if filename:
        m2 = re.search(r"[LC]_(\d{4})(\d{3,5})[A-Z]{2}\.", filename)
        if m2:
            year = m2.group(1)
            num = str(int(m2.group(2)))  # removes leading zeros (01003 -> 1003)
            return f"{year}/{num}"

    return ""
```

```python
import re

def extract_text(obj):
    if isinstance(obj, str):
        return obj
    elif isinstance(obj, dict):
        return obj.get("#text", "")
    elif isinstance(obj, list):
        return " ".join(extract_text(o) for o in obj)
    return ""

def parse_act_from_dict(content_dict: dict) -> dict:
    act = content_dict.get("ACT") or {}

    # 1. act_id
    bib = act.get("BIB.INSTANCE") or {}
    no_doc = bib.get("NO.DOC") or {}
    if isinstance(no_doc, list):
        no_doc = no_doc[0] if no_doc else {}
    year = no_doc.get("YEAR", "")
    number = no_doc.get("NO.CURRENT", "")
    act_id = f"{year}/{number}" if year and number else ""

    # 2. title, reg_number, act_type
    title_parts = act.get("TITLE", {}).get("TI", {}).get("P", [])
    if isinstance(title_parts, dict):
        title_parts = [title_parts]
    title_str = " ".join(extract_text(p) for p in title_parts).strip()
    reg_match = re.search(r'\d{4}/\d+', title_str)
    reg_number = act_id if reg_match else ""
    act_type = ""
    if title_parts:
        first = title_parts[0]
        if isinstance(first, dict) and "HT" in first:
            act_type = extract_text(first["HT"])

    # 3. date_signed
    date_elem = bib.get("DATE") or {}
    if isinstance(date_elem, list):
        date_elem = date_elem[0] if date_elem else {}
    date_signed = date_elem.get("@ISO", "")

    # 4. legal_basis
    legal_basis = []
    preamble = act.get("PREAMBLE") or {}
    gr_visa = preamble.get("GR.VISA") or {}
    gr_visa_list = gr_visa if isinstance(gr_visa, list) else [gr_visa]
    for gv in gr_visa_list:
        if not isinstance(gv, dict):
            continue
        visas = gv.get("VISA", [])
        if isinstance(visas, (dict, str)):
            visas = [visas]
        elif visas is None:
            visas = []
        for v in visas:
            legal_basis.append(extract_text(v))

    # 5. recitals
    recitals = []
    gr_consid = preamble.get("GR.CONSID") or {}
    consid_list = gr_consid.get("CONSID", [])
    if isinstance(consid_list, dict):
        consid_list = [consid_list]
    for c in consid_list:
        if not isinstance(c, dict):
            continue
        np = c.get("NP") or {}
        if isinstance(np, list):
            np = np[0] if np else {}
        txt = extract_text(np.get("TXT", "")) if isinstance(np, dict) else ""
        if txt:
            recitals.append(txt.strip())

    # 6. articles
    articles = []
    terms = act.get("ENACTING.TERMS") or {}
    raw_articles = terms.get("ARTICLE", [])
    if isinstance(raw_articles, dict):
        raw_articles = [raw_articles]
    elif raw_articles is None:
        raw_articles = []
    for a in raw_articles:
        if not isinstance(a, dict):
            continue
        num = extract_text(a.get("TI.ART", "")).replace("Article", "").strip()
        title = extract_text(a.get("STI.ART", "")) if "STI.ART" in a else ""
        body = a.get("ALINEA", "")
        if isinstance(body, dict):
            body_text = extract_text(body)
        elif isinstance(body, str):
            body_text = body
        elif isinstance(body, list):
            body_text = " ".join(extract_text(b) for b in body)
        else:
            body_text = ""
        articles.append({
            "article_number": num,
            "article_title": title,
            "article_body": body_text.strip()
        })

    # 7. signatory (Œ±ŒΩŒ∏ŒµŒ∫œÑŒπŒ∫œå œÉŒµ ŒüŒõŒïŒ£ œÑŒπœÇ œÄŒµœÅŒπœÄœÑœéœÉŒµŒπœÇ)
    signatory = []
    signature = act.get("FINAL", {}).get("SIGNATURE")
    if isinstance(signature, dict):
        sig = signature.get("SIGNATORY", [])
        if isinstance(sig, dict):
            sig = [sig]
        elif not isinstance(sig, list):
            sig = []
        for s in sig:
            if not isinstance(s, dict):
                continue
            ps = s.get("P", [])
            if isinstance(ps, (dict, str)):
                ps = [ps]
            elif ps is None:
                ps = []
            for p in ps:
                signatory.append(extract_text(p).strip())

    return {
        "act_id": act_id,
        "act_type": act_type,
        "reg_number": reg_number,
        "title": title_str,
        "date_signed": date_signed,
        "legal_basis": legal_basis,
        "recitals": recitals,
        "articles": articles,
        "signatory": signatory
    }
```

```python
def recitals_to_description(recitals):
    """
    Œ†Œ±ŒØœÅŒΩŒµŒπ œÑŒ∑ ŒªŒØœÉœÑŒ± recitals (ŒÆ None) Œ∫Œ±Œπ œÑŒ∑ ŒºŒµœÑŒ±œÑœÅŒ≠œÄŒµŒπ œÉŒµ Œ≠ŒΩŒ± œÄŒªŒÆœÅŒµœÇ Œ∫ŒµŒØŒºŒµŒΩŒø,
    œáœâœÅŒØœÇ truncation œáŒ±œÅŒ±Œ∫œÑŒÆœÅœâŒΩ.
    """
    if recitals is None:
        return ""

    if isinstance(recitals, list):
        txt = " ".join(str(r).strip() for r in recitals if r)
    else:
        txt = str(recitals)

    # Œ∫Œ±Œ∏Œ¨œÅŒπœÉŒºŒ± œÄŒøŒªŒªŒ±œÄŒªœéŒΩ Œ∫ŒµŒΩœéŒΩ
    txt = " ".join(txt.split())

    return txt
```

```python
def infer_legislation_type(act_type, title):
    """
    ŒïœÄŒπœÉœÑœÅŒ≠œÜŒµŒπ Œ≠ŒΩŒ±ŒΩ Œ±ŒΩŒ∏œÅœéœÄŒπŒΩŒø œÑœçœÄŒø œÄœÅŒ¨ŒæŒ∑œÇ.
    ŒìŒπŒ± Œ±œÅœáŒÆ, Œ∫œÅŒ±œÑŒ¨ŒºŒµ œÑŒø act_type Œ±ŒΩ œÖœÄŒ¨œÅœáŒµŒπ, Œ±ŒªŒªŒπœéœÇ œàŒ¨œáŒΩŒøœÖŒºŒµ keywords œÉœÑŒøŒΩ œÑŒØœÑŒªŒø.
    """
    if isinstance(act_type, str) and act_type.strip():
        return act_type.strip()

    text = title or ""
    for t in ["Regulation", "Directive", "Decision", "Recommendation", "Opinion"]:
        if t in text:
            return t
    return ""
```

```python
def infer_legislation_passed_by(act_type, title):
    """
    ŒëœÄŒªŒøŒØ heuristics Œ≥ŒπŒ± ŒΩŒ± Œ≤œÅŒøœçŒºŒµ œÄŒøŒπŒøœÇ ŒµŒæŒ≠Œ¥œâœÉŒµ œÑŒ∑ŒΩ œÄœÅŒ¨ŒæŒ∑.
    ŒúœÄŒøœÅŒøœçŒºŒµ ŒΩŒ± œÑŒøœÖœÇ Œ≤ŒµŒªœÑŒπœéœÉŒøœÖŒºŒµ Œ±œÅŒ≥œåœÑŒµœÅŒ±.
    """
    text = f"{act_type or ''} {title or ''}"

    # Commission-only œÄœÅŒ¨ŒæŒµŒπœÇ
    if "Commission" in text and "European Parliament" not in text and "Council" not in text:
        return "European Commission"

    # ŒöŒøŒπŒΩŒ≠œÇ œÄœÅŒ¨ŒæŒµŒπœÇ ŒïŒö + Œ£œÖŒºŒ≤ŒøœÖŒªŒØŒøœÖ
    if "European Parliament and of the Council" in text or "European Parliament and the Council" in text:
        return "European Parliament and the Council of the European Union"

    # Œ†œÅŒ¨ŒæŒµŒπœÇ ŒºœåŒΩŒø œÑŒøœÖ Œ£œÖŒºŒ≤ŒøœÖŒªŒØŒøœÖ
    if "Council Decision" in text or "Council Regulation" in text:
        return "Council of the European Union"

    # Fallback
    return "European Union"
```

```python
primary_keywords = [
    "company law","company", "companies",
    "companies law", "corporate",
    "shareholder", "shareholders",
    "corporate law",
    "company legislation",
    "corporate governance",
    "share capital",
    "statutory audit",
    "annual accounts",
    "financial statements",
    "limited liability",
    "public limited company",
    "private limited company",
    "merger","share capital", "capital company",
    "mergers",
    "takeover",
    "takeovers",
    "cross-border merger",
    "cross-border conversion",
    "cross-border division",
    "insolvency",
    "bankruptcy",
    "restructuring",
    "limited liability",
    "public limited",
    "private limited",
    "partnership",
]


secondary_keywords = [
    "company",
    "companies",
    "corporate",
    "shareholder",
    "shareholders",
    "undertaking",
    "undertakings",
    "enterprise",
    "enterprises",
]
```

```python
def method1_company_flag(text: str) -> bool:
    text = str(text).lower()

    # strong hits
    if any(kw in text for kw in primary_keywords):
        return True

    # weaker, Œ¥ŒØŒΩŒøœÖŒºŒµ Œ≠ŒΩŒ± œÄŒπŒø ‚ÄúœÉœÜŒπœáœÑœå‚Äù check Œ≥ŒπŒ± secondary
    sec_hits = sum(1 for kw in secondary_keywords if kw in text)
    return sec_hits >= 2  # œÑŒøœÖŒªŒ¨œáŒπœÉœÑŒøŒΩ 2 Œ¥ŒµœÖœÑŒµœÅŒµœçŒøœÖœÉŒµœÇ ŒªŒ≠ŒæŒµŒπœÇ

df_methods["m1_company_rule"] = df_methods["normalized_text"].apply(method1_company_flag)

print("Method 1 ‚Äì keyword rule counts:")
print(df_methods["m1_company_rule"].value_counts())
```

```python
anchored_terms = [
    "company law",
    "law on companies",
    "laws on companies",
    "law of companies",
]
anchored_terms = [t.lower() for t in anchored_terms]
```

```python
def method2_company_flag(text: str) -> bool:
    text = str(text).lower()
    return any(term in text for term in anchored_terms)

df_methods["m2_company_anchor"] = df_methods["normalized_text"].apply(method2_company_flag)

print("Method 2 ‚Äì anchored phrase counts:")
print(df_methods["m2_company_anchor"].value_counts())
```
